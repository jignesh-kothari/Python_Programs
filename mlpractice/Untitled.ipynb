```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df3 = df2.copy()
df3[-df3['total_sqft'].apply(is_float)].head(11)
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) ==2:
        return (float(tokens[0])+ float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```
